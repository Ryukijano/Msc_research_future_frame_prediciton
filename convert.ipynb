```python
import os
```

```python
def convert_ipynb_to_pdf(ipynb_file):
    """
    Convert an IPython notebook (.ipynb) file to PDF.

    Parameters:
    ipynb_file (str): Path to the input .ipynb file.
    """
    # Specify the output PDF file name
    pdf_file = ipynb_file.replace(".ipynb", ".pdf")

    # Convert the notebook to PDF using the jupyter nbconvert command
    os.system(f"jupyter nbconvert --to pdf {ipynb_file}")

    # Return the path to the generated PDF file
    return pdf_file

# Example usage
ipynb_file = "C:\\Users\\sc23gd\\Documents\\GitHub\\Msc_research_future_frame_prediciton\\CONV_lstm_implementations\\future_frame_conv_ae_implementation.ipynb"
pdf_file = convert_ipynb_to_pdf(ipynb_file)
print(f"Converted {ipynb_file} to {pdf_file}")
```
